Se va a usar un Data Set sobre cáncer de mama, la información del mismo se puede ver en 
https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, cluster
```

```python
datos = pd.read_csv('data/data.csv',index_col=0)
```

```python
pd.set_option('display.max_columns', 32) 
datos.head()
```

```python
datos.describe()
```

```python
datos = datos.drop(['Unnamed: 32'], axis = 1) # Esta columna vienen toda con NaNs
datos.info()
```

```python
datos['diagnosis'] = datos['diagnosis'].map({'M':1,'B':0}) # Se codifica eel tipo de cáncer de manera binaria
```

```python

from sklearn import preprocessing
datos_escalados = pd.DataFrame(preprocessing.scale(datos.iloc[:,1:32])) #Se dejó por fuera columna 'diagnosis'
datos_escalados.columns = list(datos.iloc[:,1:32].columns)
```

```python
datos_escalados.describe()
```

## Feature Agglomeration

```python
agglo = cluster.FeatureAgglomeration(n_clusters=2).fit_transform(datos_escalados)
color=[]
for i in datos.diagnosis:
    if i==0:
        color.append('g')
    if i==1:
        color.append('b')
    if i==2:
        color.append('r')

plt.scatter(agglo[:,0],agglo[:,1],c=color)
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter= 4000, random_state = 2)
Y = tsne.fit_transform(datos_escalados)
```

```python
plt.scatter(Y[:,0],Y[:,1],  c = datos['diagnosis'], cmap = "jet", edgecolor = "None", alpha=0.35)
```

```python
agglo = cluster.FeatureAgglomeration(n_clusters=2,affinity= 'l2', linkage = 'average' ).fit_transform(datos_escalados)


color=[]
for i in datos.diagnosis:
    if i==0:
        color.append('g')
    if i==1:
        color.append('b')
    if i==2:
        color.append('r')

plt.scatter(agglo[:,0],agglo[:,1],c=color)
```

```python
agglo = cluster.FeatureAgglomeration(n_clusters=2,affinity= 'cosine', linkage = 'complete' ).fit_transform(datos_escalados)


color=[]
for i in datos.diagnosis:
    if i==0:
        color.append('g')
    if i==1:
        color.append('b')
    if i==2:
        color.append('r')

plt.scatter(agglo[:,0],agglo[:,1],c=color)
```
